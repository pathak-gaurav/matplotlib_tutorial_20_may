### Importing lib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Reading CSV

```python
gas_prices = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.plot(gas_prices.Year,gas_prices.USA,'b.-',label='United States')
plt.plot(gas_prices.Year,gas_prices.Canada,'r.-',label='Canada')
plt.plot(gas_prices.Year,gas_prices['South Korea'],'y.-',label='South Korea')

plt.title('World Gas Prices',fontdict={'fontname':'Comic Sans MS','fontsize':'16'})
plt.xlabel('Year',fontdict={'fontname':'Comic Sans MS','fontsize':'12'})
plt.ylabel('Prices(USD)',fontdict={'fontname':'Comic Sans MS','fontsize':'12'})

plt.xticks(gas_prices.Year[::2])
plt.legend()

plt.show()
```
